```python
import ast
import numpy as np
import pandas as pd

def convert(obj):
    l=[]
    for i in ast.literal_eval(obj) :
        # ast.literal_eval works like the eval that if we give a string which
        # is of form of a dictonary then it makes it a dictinory object
        l.append(i['name'])
    return l

def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj) :
        if counter != 3 :
            l.append(i['name'])
            counter+=1
        else : 
            break
        return l

def fetchDirector(obj):
    l= []
    for i in ast.literal_eval(obj) :
        if i['job'].capitalize() == 'Director' :
            l.append(i['name'])
    return l
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    stemmed = []
    for i in text.split():
        stemmed.append(ps.stem(i))
    return " ".join(stemmed)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
vectors = cv.fit_transform(new_df['tags']).toarray()

# converts each row of the column into an array of number of times each word is repeated
# Ex : text = [â€˜Hello my name is james, this is my python notebookâ€™]
# This gives 
#  Hello    my   name    is   james   python     notebook
#    1       2     1      2     1       1          1
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```
